```python
!pip install jieba wordcloud

!wget -O NotoSansCJKtc-Regular.otf https://github.com/googlefonts/noto-cjk/blob/main/Sans/OTF/TraditionalChinese/NotoSansCJKtc-Regular.otf?raw=true


import matplotlib.font_manager as fm
font_path = 'NotoSansCJKtc-Regular.otf'
fontprop = fm.FontProperties(fname=font_path)

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = fontprop.get_name()


from google.colab import files
uploaded = files.upload()


import pandas as pd
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)


df["學校"] = df["學校"].replace("112學年度", "台大")


import jieba
import re
from collections import Counter

text_data = " ".join(df["課程名稱"].astype(str))
text_data = re.sub(r"[^\w\s]", "", text_data)
words = jieba.lcut(text_data)


stopwords = set(["的", "與", "及", "在", "是", "有", "也", "到", "和", "由", "或", "之", "並"])
filtered_words = [w for w in words if w not in stopwords and len(w) > 1]


word_freq = Counter(filtered_words)
top_words = word_freq.most_common(20)

from wordcloud import WordCloud

wordcloud = WordCloud(
    font_path="NotoSansCJKtc-Regular.otf",
    background_color="white",
    width=800,
    height=400
).generate_from_frequencies(dict(word_freq)) #Pass a dictionary to generate_from_frequencies

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("課程名稱文字雲", fontproperties=fontprop)
plt.show()

words, counts = zip(*top_words)
plt.figure(figsize=(10, 6))
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("高頻詞詞頻統計（Top 20）", fontproperties=fontprop)
plt.xlabel("詞彙", fontproperties=fontprop)
plt.ylabel("出現次數", fontproperties=fontprop)
plt.tight_layout()
plt.show()


import matplotlib.font_manager as fm
font_path = 'NotoSansCJKtc-Regular.otf'
fontprop = fm.FontProperties(fname=font_path)

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = fontprop.get_name()


from google.colab import files
uploaded = files.upload()


import pandas as pd
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)


df["學校"] = df["學校"].replace("112學年度", "台大")


import jieba
import re
from collections import Counter

text_data = " ".join(df["課程名稱"].astype(str))
text_data = re.sub(r"[^\w\s]", "", text_data)
words = jieba.lcut(text_data)


stopwords = set(["的", "與", "及", "在", "是", "有", "也", "到", "和", "由", "或", "之", "並"])
filtered_words = [w for w in words if w not in stopwords and len(w) > 1]


word_freq = Counter(filtered_words)
top_words = word_freq.most_common(20)

from wordcloud import WordCloud

wordcloud = WordCloud(
    font_path="NotoSansCJKtc-Regular.otf",
    background_color="white",
    width=800,
    height=400
).generate_from_frequencies(word_freq)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("課程名稱文字雲", fontproperties=fontprop)
plt.show()

words, counts = zip(*top_words)
plt.figure(figsize=(10, 6))
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("高頻詞詞頻統計（Top 20）", fontproperties=fontprop)
plt.xlabel("詞彙", fontproperties=fontprop)
plt.ylabel("出現次數", fontproperties=fontprop)
plt.tight_layout()
plt.show()
```
